```python
import CactusTool
import matplotlib.pyplot as plt

name = 'BBH002Lev1'
sim = CactusTool.load(name, '/Volumes/simulations/BBH_Catalog')
# sim = CactusTool.load(name, '/Users/liuyu/Desktop/Remote')
time_total = sim.Scalar(ftype='')['carpet-timing'].dsets('time_total')

fig = plt.figure()
ax = fig.add_subplot(111)

ax.set_title('time_total')
ax.set_xlabel('time [M]')
ax.set_ylabel('Day [HPC]')
ax.plot(time_total[0], time_total[1]/(60*60*24), 'k', label=name)
ax.legend()
# ax.set_xlim(0,10)
# ax.set_ylim(0,0.3)
```
