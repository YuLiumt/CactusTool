```python
import matplotlib.pyplot as plt
import numpy as np
t = strain[0]
hp = strain[1].real
hc = strain[1].imag
hp_dot = np.gradient(hp, t)
hc_dot = np.gradient(hc, t)

Omega = (hp_dot*hc - hp*hc_dot) / (2 * (hp**2 + hc**2))
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(t, Omega, 'k')
```
