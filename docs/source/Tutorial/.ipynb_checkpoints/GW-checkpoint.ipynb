```python
import CactusTool

name = 'GW150914_28'
# name = 'HMNS_long'
sim = CactusTool.load(name, basedir='/Volumes/simulations')
mp = sim.Analysis('multipole')
```

## Psi4

```python
import matplotlib.pyplot as plt
import numpy as np

mode = (2,2)
radiu = 500.0
Psi4 = mp.psi4[mode][radiu]

fig = plt.figure()
ax = fig.add_subplot(111)
plt.xlabel('t/M')
plt.ylabel(r'$Re[\Psi_4^{(2,2)}]$')
plt.plot(Psi4[0],Psi4[1],'k')
plt.show()
```

## Strain

```python
bbh = sim.Analysis('TwoPunctures')
strain = mp.Strain(bbh.ADMMass, bbh.CutoffFrequency, (2,2), -1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111)
plt.xlabel('(t-r*)/M')
plt.ylabel('Re[h+^(2,2)]')
plt.plot(strain[0],strain[1].real,'k')
# plt.ylim(-.5,.5)
# plt.xlim(0,1500)

plt.show()
```
