## Orbital Eccentricity

We computed the coordinate distance $D$ between the star centers at each time step, and fitted its derivative with the following Newtonian approximation for the orbital evolution:

$$
\dot{D}(t)=A_{0}+A_{1} t-e D_{0} \omega_{e} \sin \left(\omega_{e} t+\phi_{e}\right)
$$
where $e$ is the eccentricity and $D_{0}$ the initial coordinate interbinary distance. The fit is performed in the time interval between $t_{\mathrm{ret}}=3 \mathrm{ms}$ and $t_{\mathrm{ret}}=\frac{2}{3} t_{\mathrm{merger}}$ to avoid the initial spurious radiation and the plunge phase but having at least one eccentricity cycle included.

arXiv:1605.03424
