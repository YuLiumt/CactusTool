```python

import pandas as pd
def test():
    import CactusTool
    
    file = '/Users/liuyu/simulations/GW150914/phi.0.xy.h5'
    a = header_h5(file)
    b = select_header_h5(a, var='phi', it=0, rl=-1, c=-1)
    with read(file) as f:
        tem = []
        for item in b:
            dset = f[item]
            data = np.array(dset)
            origin = dset.attrs.get('origin', None)
            delta = dset.attrs.get('delta', None)
            size = data.shape
            dim = 2
            coord = tuple(np.arange(0,size[(dim-1)-i])*delta[i]+origin[i] for i in range(dim))
            grid = np.meshgrid(*coord)
            tem_c = pd.DataFrame()
            for i in range(dim):
                tem_c[i] = grid[i].flatten()
            tem_c['phi'] = data.flatten()
            tem_c['time'] = dset.attrs.get('time', None)
            tem.append(tem_c)
    return tem
    
```
